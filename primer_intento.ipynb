```python
import pandas as pd
```

```python
todos_centros = pd.read_excel('data/listado_todos_centros.xls', header=0)
```

```python
privados_sin = pd.read_excel('data/listado_centros_no_concertados.xls', header=0)
```

```python
todos_centros_detalle = pd.concat([todos_centros, privados_sin]).sort_values(['Provincia', 'Localidad', 'Denominación específica', 'Naturaleza']).drop_duplicates(subset=['Código'], keep='last')
```

```python
todos_centros_detalle = todos_centros_detalle.replace({"Centro privado": "Centro concertado", "Centro privado no concertado":"Centro privado"})
```
